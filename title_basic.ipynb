```python
# Import Dependancies 
import pandas as pd
```

```python
# Download File
title_basics = "title.basics.tsv"
```

```python
# Read in File title.basics.tsv
title_df_og = pd.read_csv('title.basics.tsv', sep='\t')
title_df_og.head()
```

```python
title_df = title_df_og
```

```python
# Discover datatypes 
title_df.dtypes
```

```python
title_df["startYear"].value_counts()
```

```python
title_df["endYear"].value_counts()
```

```python
# Drop endYear from DataFrame
title_df = title_df.drop("endYear", axis='columns')
title_df.head()
```

```python
title_df["titleType"].value_counts()
```

```python
# Filter DataFrame so only "movie" is the titleType
title_df = title_df[title_df['titleType'] == "movie"]
title_df
```

```python
# Drop "originalTitle" from DataFrame
title_df = title_df.drop("originalTitle", axis='columns')
title_df.head()
```

```python
# Drop "primaryTitle" from DataFrame
title_df = title_df.drop("primaryTitle", axis='columns')
title_df.head()
```

```python
title_df.dtypes
```

```python
import numpy as np
title_df = title_df.replace(to_replace="\\N",
           value=np.NaN)
```

```python
title_df
```

```python
# Drop Null values in 'runtimeMinutes'
title_df = title_df.dropna(subset=['runtimeMinutes'])
```

```python
title_df
```

```python
# Filter DataFrame so only "isAdult" is 0
title_df = title_df[title_df['isAdult'] == 0]
title_df
```

```python
# Drop 'isAdult' column
title_df = title_df.drop("isAdult", axis='columns')
title_df.head()
```

```python
# Drop 'titleType' column
title_df = title_df.drop("titleType", axis='columns')
title_df.head()
```

```python
# Drop Null values in 'startYear'
title_df = title_df.dropna(subset=['startYear'])
```

```python
title_df['startYear'].isnull().values.any()
```

```python
# Drop Null values in 'genres'
title_df = title_df.dropna(subset=['genres'])
```

```python
title_df['genres'].isnull().values.any()
```

```python
title_df.isnull().values.any()
```

```python
title_df.to_csv("title_basics2.csv", sep='\t', encoding='utf-8', index=False)
```
