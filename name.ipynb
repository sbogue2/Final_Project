```python
# Import Dependancies 
import pandas as pd
import numpy as np
```

```python
# Read in File title.principals.tsv
df = pd.read_csv('name.basics.tsv', sep='\t')
df.head()
```

```python
name_df = df
```

```python
# Discover data types
name_df.dtypes
```

```python
# Drop "birthYear" from DataFrame
name_df = name_df.drop("birthYear", axis='columns')
name_df.head()
```

```python
# Drop "deathYear" from DataFrame
name_df = name_df.drop("deathYear", axis='columns')
name_df.head()
```

```python
# Drop "primaryProfession" from DataFrame
name_df = name_df.drop("primaryProfession", axis='columns')
name_df.head()
```

```python
# Replace "/N" value with "NaN"
name_df = name_df.replace(to_replace="\\N",
           value=np.NaN)
```

```python
# Check for null values
name_df.isnull().values.any()
```

```python
# Remove null values from DataFrame
name_df = name_df.dropna()
```

```python
# Check for null values
name_df.isnull().values.any()
```

```python
name_df
```

```python
# Export principals_df DataFrame as csv file
name_df.to_csv("name_df.csv", sep='\t', encoding='utf-8', index=False)
```
