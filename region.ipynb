```python
# Import Dependancies 
import pandas as pd
import numpy as np
```

```python
# Read in File title.principals.tsv
df = pd.read_csv('title.akas.tsv', sep='\t')
df.head()
```

```python
region_df = df
```

```python
# Drop "ordering" from DataFrame
region_df = region_df.drop("ordering", axis='columns')
region_df.head()
```

```python
# Drop "title" from DataFrame
region_df = region_df.drop("title", axis='columns')
region_df.head()
```

```python
# Drop "language" from DataFrame
region_df = region_df.drop("language", axis='columns')
region_df.head()
```

```python
# Drop "attributes" from DataFrame
region_df = region_df.drop("attributes", axis='columns')
region_df.head()
```

```python
# Drop "isOriginalTitle" from DataFrame
region_df = region_df.drop("isOriginalTitle", axis='columns')
region_df.head()
```

```python
# Drop "types" from DataFrame
region_df = region_df.drop("types", axis='columns')
region_df.head()
```

```python
# Replace "/N" value with "NaN"
region_df = region_df.replace(to_replace="\\N",
           value=np.NaN)
```

```python
# Check for null values
region_df.isnull().values.any()
```

```python
# Remove null values from DataFrame
region_df = region_df.dropna()
```

```python
# Check for null values
region_df.isnull().values.any()
```

```python
# Export principals_df DataFrame as csv file
region_df.to_csv("region_df.csv", sep='\t', encoding='utf-8', index=False)
```
